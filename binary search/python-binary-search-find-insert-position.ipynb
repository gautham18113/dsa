### State the solution in plain english and implement the same

There are two cases for this problem

**Element present in array**

Use binary search to find the element.

**Element not present in array**

Use binary search to search for target. 

To find the index where element should be inserted if it is not found, keep a track of the most recent index whose element was greater than the target.

Then answer is that index minus 1.

If there is no such index, then answer is last index + 1.

If maximum element is at 0, then insert at 0.

```python
def find_or_insert(nums, target):
    if not nums:
        return -1
    
    lo = 0
    hi = len(nums) - 1
    
    ins_idx = -1
    
    while(hi >= lo):
        
        mid = (lo + hi) // 2
        
        print("lo " + str(lo) + " hi " + str(hi) + " mid " + str(mid) + " mid num " + str(nums[mid]) + " ins idx " + str(ins_idx))
        
        if nums[mid] == target:
            return mid
        elif nums[mid] > target:
            ins_idx = mid
            hi = mid - 1
        else:
            lo = mid + 1

    if ins_idx == -1 and target < nums[0]:
        ins_idx = 0
    elif ins_idx == -1 and target > nums[len(nums)-1]:
        ins_idx = len(nums)

    return ins_idx
        
      
```

### Analyze the algoritm and identify inefficiencies if any

Using a variable to track position of element greater than target is not needed, since this number is already kept track of in the `lo` variable.

That is, if element is not present - `lo` will always have the location where the number is supposed to be. 

```python
def find_or_insert_optimized(nums, target):
    
    if not nums:
        return -1

    lo = 0
    hi = len(nums) - 1
    
    while(hi >= lo):
        
        mid = (lo + hi) // 2
        
        ## print("lo " + str(lo) + " hi " + str(hi) + " mid " + str(mid) + " mid num " + str(nums[mid]) + " ins idx " + str(ins_idx))
        
        if nums[mid] == target:
            return mid
        elif nums[mid] > target:
            hi = mid - 1
        else:
            lo = mid + 1


    return lo
```
